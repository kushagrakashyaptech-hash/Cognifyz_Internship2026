## Feature Engineering

### Objective

Extract additional features from existing columns and create new encoded features for machine learning and analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")

df.head()
```

```python
df = df.dropna(subset=['Cuisines'])

print(df.shape)
```

```python
df["Restaurant Name Length"] = df["Restaurant Name"].str.len()

df[["Restaurant Name","Restaurant Name Length"]].head()
```

```python
df["Address Length"] = df["Address"].str.len()

df[["Address","Address Length"]].head()
```

```python
df["Table Booking Encoded"] = (
    df["Has Table booking"]
    .map({"Yes":1,"No":0})
)

df[
    ["Has Table booking",
     "Table Booking Encoded"]
].head()
```

```python
df["Online Delivery Encoded"] = (
    df["Has Online delivery"]
    .map({"Yes":1,"No":0})
)

df[
    ["Has Online delivery",
     "Online Delivery Encoded"]
].head()
```

```python
new_features = [
    "Restaurant Name Length",
    "Address Length",
    "Table Booking Encoded",
    "Online Delivery Encoded"
]

df[new_features].head()
```

```python
print(df[new_features].describe())
```
