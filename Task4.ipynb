## Table Booking and Online Delivery Analysis

### Objective

Determine the percentage of restaurants that offer table booking and online delivery, compare average ratings based on table booking availability, and analyze online delivery availability across different price ranges.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")

df.head()
```

```python
df = df.dropna(subset=['Cuisines'])

print(df.shape)
```

```python
df = df.dropna(subset=['Cuisines'])

print(df.shape)
```

```python
table_booking_percentage = (
    df["Has Table booking"]
    .value_counts(normalize=True)
    * 100
)

print(table_booking_percentage)
```

```python
table_booking_percentage.plot(kind='bar')

plt.title("Percentage of Restaurants Offering Table Booking")
plt.ylabel("Percentage")

plt.show()
```

```python
online_delivery_percentage = (
    df["Has Online delivery"]
    .value_counts(normalize=True)
    * 100
)

print(online_delivery_percentage)
```

```python
online_delivery_percentage.plot(kind='bar')

plt.title("Percentage of Restaurants Offering Online Delivery")
plt.ylabel("Percentage")

plt.show()
```

```python
average_rating = (
    df.groupby("Has Table booking")
    ["Aggregate rating"]
    .mean()
)

print(average_rating)
```

```python
average_rating.plot(kind='bar')

plt.title("Average Rating by Table Booking Availability")
plt.ylabel("Average Rating")

plt.show()
```

```python
price_delivery = pd.crosstab(
    df["Price range"],
    df["Has Online delivery"]
)

print(price_delivery)
```

```python
price_delivery.plot(kind='bar', figsize=(8,5))

plt.title("Online Delivery Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.show()
```
