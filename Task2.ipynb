## Descriptive Analysis

### Objective

Calculate basic statistical measures for numerical columns, explore the distribution of categorical variables, and identify the top cuisines and cities with the highest number of restaurants.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df = df.dropna(subset=['Cuisines'])
print(df.isnull().sum())
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64'])

numerical_columns.columns
```

```python
numerical_columns.describe()
```

```python
df["Country Code"].value_counts()
```

```python
df["Country Code"].value_counts().plot(kind='bar')
plt.title("Distribution of Country Codes")
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
top_cities = df["City"].value_counts()
top_cities.head(10)
```

```python
top_cities.head(10).plot(kind='bar')

plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
top_cuisines = df["Cuisines"].value_counts()

top_cuisines.head(10)
```

```python
top_cuisines.head(10).plot(kind='bar')

plt.title("Top 10 Cuisine Categories")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=45)

plt.show()
```
