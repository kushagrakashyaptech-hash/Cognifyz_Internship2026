## Data Visualization

### Objective

Create visualizations to analyze rating distributions, compare average ratings across cuisines and cities, and study relationships between restaurant features and ratings.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df = df.dropna(subset=['Cuisines'])

print(df.shape)
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["Aggregate rating"],
    bins=20
)

plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")

plt.show()
```

```python
rating_counts = (
    df["Aggregate rating"]
    .round()
    .value_counts()
    .sort_index()
)

rating_counts.plot(
    kind="bar"
)

plt.title("Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
top_cuisines = (
    df.groupby("Cuisines")
    ["Aggregate rating"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print(top_cuisines)
```

```python
top_cuisines.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Cuisines by Average Rating")
plt.xlabel("Cuisine")
plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.show()
```

```python
top_cities = (
    df.groupby("City")
    ["Aggregate rating"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print(top_cities)
```

```python
top_cities.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Votes"],
    df["Aggregate rating"],
    alpha=0.5
)

plt.title("Votes vs Aggregate Rating")
plt.xlabel("Votes")
plt.ylabel("Aggregate Rating")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Average Cost for two"],
    df["Aggregate rating"],
    alpha=0.5
)

plt.title("Average Cost for Two vs Rating")
plt.xlabel("Average Cost for Two")
plt.ylabel("Aggregate Rating")

plt.show()
```

```python
price_rating = (
    df.groupby("Price range")
    ["Aggregate rating"]
    .mean()
)

print(price_rating)
```

```python
price_rating.plot(
    kind="bar"
)

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

plt.show()
```
