## Data Exploration and Preprocessing

### Objective
Explore the dataset, identify missing values, analyze data types, and study the distribution of the target variable (Aggregate Rating).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Cuisines'])

print(df.isnull().sum())
```

```python
df.dtypes
```

```python
df["Aggregate rating"].describe()
```

```python
df["Aggregate rating"].value_counts().sort_index()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Aggregate rating"], bins=20)
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```
